<h1 style="font-size:20px;">Summary of the Data Set</h1>
</center>


This dataset contains the nutrition info for Starbucks menu items.

1. The 'Beverage_category' column classifies the type of beverage, such as coffee, tea, or smoothie. 
2. The 'Beverage' column provides the specific name of the drink, for instance, Caramel Macchiato or Green Tea Latte.

3. The 'Beverage_prep' column details the preparation method of the beverage, including whether it's served hot or cold, and any additional ingredients or toppings like whipped cream or syrup. 
4. The 'Calories' column lists the total caloric content of each beverage, providing insight into the energy provided by each drink.

5. The next three columns, 'Total Fat (g)', 'Trans Fat (g)', and 'Saturated Fat (g)', provide a breakdown of the fat content in each beverage. These columns are crucial for those monitoring their fat intake for health or dietary reasons.
6. The 'Sodium (mg)' column indicates the amount of sodium in each beverage, which is essential information for individuals on low-sodium diets.

7. The 'Total Carbohydrates (g)' column provides the total carbohydrate content, including sugars, which is particularly useful for people managing diabetes or following a low-carb diet.
8. Lastly, the 'Cholesterol (mg)' column lists the amount of cholesterol in each beverage, a critical factor for those monitoring their cholesterol levels.

This dataset serves as a comprehensive guide to the nutritional content of Starbucks beverages, making it a valuable resource for researchers, dietitians, and health-conscious consumers.

```python
sns.countplot(data= starbucks, y = 'Beverage_category', palette= 'viridis')
plt.title("Count of Beverage Category")
```

```python
plt.figure(figsize=(10, 6))
starbucks.groupby('Beverage_category')['Calories'].mean().plot(kind='barh', color=sns.color_palette("Set2"))
plt.title('Average Calories by Beverage Category')
plt.ylabel('Beverage Category')
plt.xlabel('Average Calories')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
starbucks.groupby('Beverage_category')[' Sugars (g)'].mean().plot(kind='barh', color=sns.color_palette("Set2"))
plt.title('Average Sugar Content by Beverage Category')
plt.ylabel('Beverage Category')
plt.xlabel('Average Sugar Content')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
starbucks.groupby('Beverage_category')['Caffeine (mg)'].mean().plot(kind='barh', color=sns.color_palette("Set2"))
plt.title('Average Caffeine(mg) by Beverage Category')
plt.ylabel('Beverage Category')
plt.xlabel('Average Caffeine (mg)')
plt.show()
```

```python
# Create a new DataFrame with the average fat values for each beverage category
fat_columns = [' Total Fat (g)', 'Trans Fat (g) ', 'Saturated Fat (g)']
fat_data = starbucks.groupby('Beverage_category')[fat_columns].mean()

# Plotting the bar chart
fat_data.plot(kind='bar', figsize=(12, 6), color=sns.color_palette("Set2"))

# Adding titles and labels
plt.title('Comparison of Total Fat, Trans Fat, and Saturated Fat by Beverage Category')
plt.xlabel('Beverage Category')
plt.ylabel('Fat (g)')
plt.xticks(rotation=45)
plt.legend(title='Fat Type', bbox_to_anchor=(1, 1), loc='upper left')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group by 'Beverage' and calculate the average cholesterol content
top_cholesterol_beverages = starbucks.groupby('Beverage')['Cholesterol (mg)'].mean()

# Sort by cholesterol content in descending order and get the top 10
top_cholesterol_beverages = top_cholesterol_beverages.sort_values(ascending=True).head(5)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
top_cholesterol_beverages.plot(kind='barh', color=sns.color_palette("viridis", n_colors=10))

# Adding titles and labels
plt.title('Best 5 Beverages Based on Least Cholesterol Content')
plt.xlabel('Average Cholesterol (mg)')
plt.ylabel('Beverage')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group by 'Beverage' and calculate the average protein content
top_protein_beverages = starbucks.groupby('Beverage')[' Protein (g) '].mean()

# Sort by protein content in descending order and get the top 5
top_protein_beverages = top_protein_beverages.sort_values(ascending=False).head(5)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
top_protein_beverages.plot(kind='bar', color=sns.color_palette("Blues", n_colors=5))

# Adding titles and labels
plt.title('Top 5 Beverages Based on Average Protein Content')
plt.xlabel('Beverage')
plt.ylabel('Average Protein (g)')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group by 'Beverage' and calculate the average caffeine content
least_caffeine_beverages = starbucks.groupby('Beverage')['Caffeine (mg)'].mean()

# Sort by caffeine content in ascending order (least caffeine first) and get the top 10
least_caffeine_beverages = least_caffeine_beverages.sort_values(ascending=True).head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
least_caffeine_beverages.plot(kind='bar', color=sns.color_palette("coolwarm", n_colors=5))

# Adding titles and labels
plt.title('Top 10 Beverages Based on Least Average Caffeine Content')
plt.xlabel('Beverage')
plt.ylabel('Average Caffeine (mg)')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
